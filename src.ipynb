```python
import numpy as np
import matplotlib.pylab as plt
import scipy.stats as s
```

```python
coin_tosses = np.random.binomial(n=5, p=0.5, size=10000000)
print(coin_tosses)
frequency = np.count_nonzero(coin_tosses == 2)
relative_frequency = frequency/coin_tosses.shape[0]
print(relative_frequency)
```

```python
coin_tosses = np.random.binomial(n=5, p=0.5, size=10000000)
outcomes= np.arange(10)
counts = np.array([np.count_nonzero(coin_tosses == k) for k in outcomes])
probabilities = counts / coin_tosses.size

plt.bar(outcomes, probabilities)
plt.xlabel("Number of desired outcomes out of 10 Bernoulli Trials")
plt.ylabel("P(K/N)")
plt.title("Binomial Distribution PMF")
plt.show()
```
